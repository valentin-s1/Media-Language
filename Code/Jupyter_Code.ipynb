# Exploratory Project – Media Framing and Public Preferences in Swiss Initiatives

## Introduction

This exploratory project investigates whether media coverage in Switzerland reflects public sentiment ahead of direct democratic referenda. We analyze whether the **tone**, **vocabulary**, and **topical framing** of newspaper articles can serve as a proxy for the public's voting preferences.

The project is conceptually situated between two perspectives on media and democracy. **Framing theory** suggests that media coverage can shape public perception by emphasizing specific narratives or problem frames. In contrast, the **mirror theory** posits that the media simply reflect pre-existing societal attitudes. We explore whether the linguistic content of media reports aligns with actual vote outcomes, as a possible indication of representativeness.

The dataset comprises **113 articles** on **20 national referenda** held between **2011 and 2024**, collected from the major Swiss outlets **SRF**, **20 Minuten**, and **NZZ**. All articles were published **before** the respective votes to avoid hindsight bias and were randomly selected to avoid bias.

The relatively **small sample size** reflects both **institutional and methodological constraints**. Since the Swiss system holds only a limited number of popular votes in a given period, and because our analysis requires articles that explicitly and centrally deal with a specific initiative (not just general political discourse), the usual broad scraping methods (e.g., by keyword or topic) were not viable. Instead, we relied on **manual article selection and validation** to ensure relevance and timing. We chose the 2011–2024 window to maximize the availability of digitized media content.

Below, we define a custom scraping function that takes a URL and extracts the article content. It uses the `requests` library to access the page and `BeautifulSoup` to parse the HTML. The function retrieves all visible paragraph tags (`<p>`) and concatenates them into a single text string. We then apply this function only to articles from SRF and 20 Minuten, for which no text has yet been collected. The results are saved in the *text* column and exported to a filled Excel file.


```python
# Function to scrape URL and extract text
def extract_text_from_url(url):
    try:
        headers = {
            "User-Agent": "Mozilla/5.0"
        }
        response = requests.get(url, headers=headers, timeout=10)
        if response.status_code != 200:
            return None
        soup = BeautifulSoup(response.content, "html.parser")
        paragraphs = soup.find_all("p")
        text = " ".join(p.get_text().strip() for p in paragraphs)
        return text if text else None
    except Exception as e:
        print(f"Fehler bei URL: {url} → {e}")
        return None

# Filter: only Articles NOT from nzz 
df_scrape = df[
    (~df["newspaper_name"].str.lower().str.contains("nzz", na=False)) &
    (~df["news_article_URL"].isnull()) &
    (df["text"].isnull() | (df["text"].str.strip() == ""))
]

# start scraping and add to existing 'text'-variable
for idx in tqdm(df_scrape.index):
    url = df.at[idx, "news_article_URL"]
    text = extract_text_from_url(url)
    df.at[idx, "text"] = text

# create new file with filled text variable 
output_path = "../Data/Data_ML_filled.xlsx"
df.to_excel(output_path, index=False)
print(f" Done, saved as: {output_path}")



```

Upon closer inspection of the raw article texts, we identified systematic non-content elements in both SRF and 20 Minuten articles. These include login instructions, app banners, author lines, and embedded lists of unrelated games or features that frequently appear at the beginning or end of the articles. As these components are not part of the actual journalistic content, we exclude them from the dataset in the following step using tailored cleaning functions for each news outlet.


```python
# Clean SRF articles by removing known junk phrases and footer blocks
def clean_srf_text(text):
    if not isinstance(text, str):
        return text

    # 1. Remove known static junk phrases
    garbage_phrases = [
        "Einige Elemente auf SRF.ch funktionieren nur mit aktiviertem JavaScript",
        "SRF News App Bei Google Play",
        "In der App öffnen", 
        "Wenn Sie sich erneut für die Kommentarfunktion registrieren möchten, melden Sie sich bitte beim Kundendienst von SRF.",
        "SRF Schweizer Radio und Fernsehen,Zweigniederlassung der Schweizerischen Radio- und Fernsehgesellschaft"

    ]
    for phrase in garbage_phrases:
        if phrase in text:
            text = text.replace(phrase, "")

    # 2. Remove dynamic noisy patterns (author info, footer, etc.)
    text = re.sub(r"Autor:\s*\n?.*?\d{2}\.\d{2}\.\d{4}, \d{2}:\d{2}", "", text)
    text = re.sub(r"Für die Registrierung benötigen wir.*", "", text)
    text = re.sub(r"Bitte melden Sie sich an, um einen Kommentar zu erfassen.*", "", text)


    return text.strip()


# Apply SRF cleaning to all SRF articles
mask_srf = df_new["newspaper_name"].str.lower().str.contains("srf", na=False)
df_new.loc[mask_srf, "text"] = df_new.loc[mask_srf, "text"].apply(clean_srf_text)


# Clean 20min articles by removing game lists and footer elements
def clean_20min_text(text):
    if not isinstance(text, str):
        return text
    
    garbage_words_20min = [
    "Deine Meinung zählt", "Solitaire", "Kreuzworträtsel", "Sudoku", "Mahjong", "Bubbles",
    "Snake", "Schach", "eXchange", "Power of 2", "Doppel", "Cuboro", "Riddles", "Wortblitz",
    "SudoKen", "Street", "Fibonacci", "Gumblast", "Rushtower", "Skiracer", "Wimmlbid", "(sda)"
]
    
    for phrase in garbage_words_20min:
        text = text.replace(phrase, "")

    return text.strip()

# Apply 20min cleaning to all 20 Minuten articles
mask_20min = df_new["newspaper_name"].str.lower().str.contains("20", na=False)
df_new.loc[mask_20min, "text"] = df_new.loc[mask_20min, "text"].apply(clean_20min_text)


## final clean: remove NaN values from the text column
df_new = df_new[df_new["text"].notnull() & (df_new["text"].str.strip() != "")]

```

### 1.2 NLP Preprocessing

Tghe text preprocessing function (***preprocess_text***) takes a raw German text input and returns a fully cleaned, tokenized, compound-split, stop-word filtered and lemmatized string. It implements best practices for preparing a corpus for classical ML models (e.g. TF-IDF) or modern embeddings.

1. **Force string & lowercase**  
   - Ensures every input is treated as text (no floats / NaNs).  
   - Lowercases to collapse “Haus” and “haus” into the same token.

2. **Remove URLs, digits & unwanted punctuation**  
   - URLs and numbers rarely carry semantic weight in political or news analysis.  
   - We strip all characters except letters, spaces and German umlauts/ß to simplify downstream tokenization.

3. **Tokenization & stop-word filtering**  
   - spaCy splits the cleaned string into tokens.  
   - Tokens marked as stop-words (e.g. “und”, “der”) or non-alphabetical are discarded to focus on content words.

4. **Compound splitting**  
   - German compounding (“Militärwaffen”) creates very long tokens, leading to data sparsity.  
   - We use a compound splitter to break each token into its meaningful parts (→ “Militär”, “Waffen”).

5. **Lemmatization**  
   - Each split piece is fed back into spaCy to obtain its lemma (dictionary form).  
   - Lemmas of length ≤2 are discarded to remove noise (e.g. “im”, “am”).

6. **Re-join tokens**  
   - The final output is a space-joined string of clean, meaningful lemmas ready for vectorization.

**Why this matters**  
- **Data quality**: eliminates noise (junk characters, filler words).  
- **Vocabulary control**: splits compounds and lemmatizes to reduce the total number of unique tokens.  
- **Model performance**: cleaner, more consistent inputs improve both statistical models (TF-IDF, bag-of-words) and neural embeddings (BERT).  


```python
## requirements: !python -m spacy download de_core_news_md
import re
import spacy
from split_words import Splitter  # your splitter
import pandas as pd

# — Load the medium model with only the components you need —
nlp = spacy.load("de_core_news_md", disable=["parser", "ner"])

# ─── Whitelist "eu" so it won't be dropped as a stop‐word or by length filter ──
if "eu" in nlp.Defaults.stop_words:
    nlp.Defaults.stop_words.remove("eu")
nlp.vocab["eu"].is_stop = False

splitter = Splitter()

# toggle splitting here
SPLIT_ENABLED = True
THRESHOLD     = 0.8  # only used if SPLIT_ENABLED is True

def preprocess_with_splits(text: str):
    # 0) Force string & basic cleanup
    text = "" if text is None else str(text).lower()
    text = re.sub(r"http\S+|www\.\S+", "", text)
    text = re.sub(r"\d+", "", text)
    text = re.sub(r"[^\w\säöüß]", " ", text)
    text = re.sub(r"\s+", " ", text).strip()

    doc         = nlp(text)
    tokens      = []
    used_splits = []

    for token in doc:
        if not token.is_alpha or token.is_stop:
            continue
        word = token.text  # already lowercase

        # ───────────────────────────── splitting ──────────────────────────────
        if SPLIT_ENABLED:
            raw_parts = splitter.split_compound(word)
            if raw_parts:
                score, p1, p2 = max(raw_parts, key=lambda x: x[0])
                if score >= THRESHOLD:
                    parts = [p1.lower(), p2.lower()]
                    used_splits.append(
                        {"orig": word, "score": score, "parts": (p1, p2)}
                    )
                else:
                    parts = [word]
            else:
                parts = [word]
        else:
            # splitting disabled
            parts = [word]

        # ───────────────────────── lemmatise every piece ──────────────────────
        for part in parts:
            lemma = nlp(part)[0].lemma_.lower()
            # keep EU even though it's length 2
            if len(lemma) > 2 or lemma == "eu":
                tokens.append(lemma)

    clean_text = " ".join(tokens)
    return clean_text, used_splits

# Apply and expand into two columns
df_new[["clean_text", "split_info"]] = (
    df_new["text"]
    .apply(preprocess_with_splits)
    .apply(pd.Series)
)
```

### 2.3 Wordclouds by Voting Outcome

To explore whether articles differ in their language depending on the outcome of a vote, we generate two wordclouds: one for initiatives that were accepted and one for those that were rejected. This allows for a visual comparison of the most frequent terms in media coverage, highlighting differences in topic emphasis, tone, or framing across accepted and rejected proposals.

While wordclouds are not precise analytical tools, they provide a useful exploratory perspective. By highlighting the most frequent terms in a large collection of texts, they can offer initial insights into framing tendencies and content focus, serving as a qualitative prelude to more formal NLP methods.

To generate the word clouds, we first tokenize the cleaned texts separately for accepted ("yes") and rejected ("no") initiatives. We then count the token frequencies and normalize them to obtain relative frequencies. By computing the difference in relative frequencies between both groups, we identify words that are disproportionately associated with either outcome. The top 100 most biased words in each direction are selected and visualized using word clouds to highlight distinctive vocabulary patterns.

```python
import matplotlib.pyplot as plt
from collections import Counter
from wordcloud import WordCloud

# 1) Collect tokens (as before)
yes_tokens = [
    tok for doc in df_new[df_new["result"] == "Angenommen"]["clean_text"].str.split()
    for tok in doc
]
no_tokens  = [
    tok for doc in df_new[df_new["result"] == "Abgelehnt"]["clean_text"].str.split()
    for tok in doc
]

# 2) Build raw counters and relative freqs
cnt_yes = Counter(yes_tokens)
cnt_no  = Counter(no_tokens)
total_yes = sum(cnt_yes.values())
total_no  = sum(cnt_no.values())

rel_yes = {w: cnt_yes[w]/total_yes for w in cnt_yes}
rel_no  = {w: cnt_no[w]/total_no  for w in cnt_no}

# 3) Compute differential scores
#   positive means “more in yes”, negative means “more in no”
diff = { 
    w: (rel_yes.get(w,0) - rel_no.get(w,0)) 
    for w in set(rel_yes) | set(rel_no)
}

# 4) Pick top-N by diff for “yes”, and top-N by -diff for “no”
N = 100

# Words where diff is largest → most “yes-biased”
yes_words = sorted(diff, key=lambda w: diff[w], reverse=True)[:N]

# Words where diff is most negative → most “no-biased”
no_words  = sorted(diff, key=lambda w: diff[w])[:N]

# 5) Build frequency dicts from their own *relative* freqs
freq_yes = {w: rel_yes.get(w,0) for w in yes_words}
freq_no  = {w: rel_no.get(w,0)  for w in no_words}

# 6) Draw side-by-side with different colormaps
wc_yes = WordCloud(width=800, height=400, background_color="white",
                   colormap="Blues").generate_from_frequencies(freq_yes)
wc_no  = WordCloud(width=800, height=400, background_color="white",
                   colormap="Oranges").generate_from_frequencies(freq_no)

fig, axes = plt.subplots(1, 2, figsize=(16,6))
axes[0].imshow(wc_yes, interpolation="bilinear")
axes[0].axis("off")
axes[0].set_title("Top 100 ‘yes-biased’ words", fontsize=16)

axes[1].imshow(wc_no, interpolation="bilinear")
axes[1].axis("off")
axes[1].set_title("Top 100 ‘no-biased’ words", fontsize=16)

plt.tight_layout()
plt.show()
```

### 2.4 Zero-Shot Topic Classification using a Pre-trained Multilingual BART Model

In this section, we aim to understand the thematic focus of media coverage across different initiatives. Specifically, we examine whether certain **topics** appear more frequently in news articles discussing **accepted** versus **rejected** initiatives.

Given the **limited size of our dataset** (~109 articles), traditional topic modeling techniques like LDA or clustering would likely perform poorly and result in noisy, uninterpretable outputs. To address this, we apply a more suitable and robust alternative: **zero-shot topic classification** using a pre-trained transformer model.

We use the `facebook/bart-large-mnli` model provided by HuggingFace, a powerful **multilingual model** trained on Natural Language Inference (NLI). It can determine whether a given text semantically fits into a label, even without prior exposure to the label during training — hence the term *zero-shot*.

What makes this method particularly advantageous for our case:
- **No training required**: The model can classify unseen text into custom-defined categories.
- **Multilingual support**: Despite being trained primarily in English, the model performs well on German texts due to its robust multilingual architecture.
- **Custom labels**: We can define our own relevant policy labels, such as "Wirtschaft" (economy), "Umwelt" (environment), "Migration", and "Sicherheit" (security).

Technically, the model receives a news article and a set of candidate labels. It returns a **score between 0 and 1** for each label, indicating how well the article matches the theme. We apply this to all articles and then aggregate the results by **initiative outcome** (accepted vs. rejected).

This approach provides a much **cleaner and more interpretable topic perspective** than unsupervised models — making it ideal for our small dataset.


```python
import os
import pandas as pd
from transformers import pipeline
from tqdm import tqdm
import torch

CACHE_FILE = "../Data/data_classified.parquet"
candidate_labels = [
    "Wirtschaft", "Umwelt", "Migration", "Gesundheit",
    "Sicherheit", "Soziales", "Bildung", "Steuern"
]

if os.path.exists(CACHE_FILE):
    # Fast‐path: load everything, no classification needed
    df_new = pd.read_parquet(CACHE_FILE)
else:
    # 1) load your raw data
    df_new = pd.read_excel("../Data/Data_ML_cleaned.xlsx")

    # 2) init the classifier (use GPU if available)
    device = 0 if torch.cuda.is_available() else -1
    classifier = pipeline(
        "zero-shot-classification",
        model="facebook/bart-large-mnli",
        device=device,
        tokenizer_kwargs={"use_fast": True},
    )

    texts = df_new["text"].fillna("").tolist()
    batch_size = 16
    results = []

    # 3) chunk & classify with a tqdm bar
    for i in tqdm(range(0, len(texts), batch_size), desc="Zero-shot batches"):
        batch = texts[i : i + batch_size]
        res = classifier(batch, candidate_labels, multi_label=True)
        results.extend(res)

    # 4) unpack into your DataFrame
    df_new["top_topic"]   = [r["labels"][0] for r in results]
    df_new["topic_score"] = [r["scores"][0] for r in results]

    # 5) cache to disk
    df_new.to_parquet(CACHE_FILE, index=False)

### Results are stored in ../Data/data_classified.parquet --> run time over 1h
```

### 3.1 Sentiment Analysis

In this section, we perform sentiment classification on the article texts using a transformer-based language model specifically fine-tuned for German sentiment detection (`oliverguhr/german-sentiment-bert`). The model outputs a categorical label (`positive`, `neutral`, or `negative`) along with a confidence score for each prediction.

We apply the model to each article by feeding the text directly into the pipeline. Since transformer models have a maximum input length of 512 tokens (typically around 4000–5000 characters for German text), we truncate longer articles to this limit. As shown in the previous figure, the majority of articles fall within this range: over 80% of texts are shorter than 5000 characters and thus remain unaffected by this constraint. Only a small number of outliers, i.e. very long articles, will be partially truncated. This trade-off allows efficient processing while preserving most of the sentiment-relevant content.

The predicted sentiment labels are then mapped to numeric values (`+1` for positive, `0` for neutral, and `–1` for negative) to enable comparative analysis across outcome groups (`accepted` vs. `rejected`). We additionally record the model’s confidence score, which serves as a proxy for classification certainty.

This setup enables us to explore whether articles covering accepted initiatives tend to be more positively framed than those covering rejected ones—offering a first empirical test of media alignment with public vote outcomes.


```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline


# load the sentiment analysis model and tokenizer
# This model is a German BERT model fine-tuned for sentiment analysis
model_name = "oliverguhr/german-sentiment-bert"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)
sentiment_pipeline = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)

# estimate sentiment for each article and store results in new columns
# Define a function to get sentiment label and score
def get_sentiment_info(text):
    result = sentiment_pipeline(text[:512])[0]  # reduction of max length to 512 tokens
    return pd.Series([result["label"], result["score"]])

df_new[["sentiment_label", "sentiment_score"]] = df_new["text"].apply(get_sentiment_info)

# add a numerical value for the sentiment label
label_map = {"positive": 1, "neutral": 0, "negative": -1}
df_new["sentiment_value"] = df_new["sentiment_label"].map(label_map)
```

#### 3.2.3 Load the tokenizer and model

We use a pre-trained German BERT model (`bert-base-german-cased`) as the base architecture for our classification task. BERT is well-suited for this purpose because it captures syntactic structure, word meaning, and contextual relationships through its bidirectional training approach.

The tokenizer converts input texts into the format expected by BERT, while the model is configured for sequence classification with three output labels: *pro*, *neutral*, and *contra*.

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification

model_name = "bert-base-german-cased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=3) # num_labels=3 for pro, neutral, contra
```

#### 3.2.4 Tokenize the Input Text

BERT requires tokenized input, meaning words must be converted into numerical IDs using a tokenizer. We truncate each input to a maximum of 512 tokens (BERT’s limit), and apply padding so that all sequences have equal length. 


Unlike traditional models, BERT does **not** require lowercasing, lemmatization, or stop word removal. This is because the tokenizer and model handle linguistic context internally, rather than treating text as a bag-of-words.

The following function applies tokenization to the input texts in our dataset:

```python
def tokenize_function(example):
    return tokenizer(example["text_input"], truncation=True, padding="max_length", max_length=512)

tokenized_datasets = dataset.map(tokenize_function, batched=True)
tokenized_datasets = tokenized_datasets.rename_column("stance_label", "labels")

```

#### 3.2.5 Define training setup and run training

We configure and launch fine-tuning of the pre-trained BERT model using HuggingFace’s `Trainer` API. The training loop uses accuracy and macro-F1 as evaluation metrics and runs for five epochs. Tokenized datasets, evaluation logic, and training hyperparameters are passed to the `Trainer`, which handles optimization and evaluation internally.

```python

#%pip install -U "transformers[torch]" "accelerate>=0.26.0" evaluate transformers
import numpy as np
import evaluate
import sys
import transformers
import accelerate
print(accelerate.__version__)  # should be >= 0.26.0

from transformers import AutoModelForSequenceClassification, AutoTokenizer, Trainer, TrainingArguments

# Load the model and tokenizer
# Load evaluation metrics
accuracy = evaluate.load("accuracy")
f1 = evaluate.load("f1")

# Define evaluation logic
def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = np.argmax(logits, axis=-1)
    return {
        "accuracy": accuracy.compute(predictions=preds, references=labels)["accuracy"],
        "f1_macro": f1.compute(predictions=preds, references=labels, average="macro")["f1"]
    }

# Define training configuration
training_args = TrainingArguments(
    output_dir="../Data/stance_model",
    eval_strategy="epoch",
    logging_dir="../Data/logs",
    num_train_epochs=5,
    learning_rate=2e-5,
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    weight_decay=0.01
)

# Launch training
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets["train"],
    eval_dataset=tokenized_datasets["test"],
    tokenizer=tokenizer,
    compute_metrics=compute_metrics
)

trainer.train()
```

**Interpretation**

The fine-tuned BERT model was trained on the manually labeled stance dataset for five epochs.  
As shown in the results, the model fails to learn meaningful class distinctions across the three stance categories (`pro`, `neutral`, `contra`):

- **Accuracy** remains constant at 0.50 across all epochs, suggesting that the model might be defaulting to a single dominant class.

- **Validation loss** shows no consistent downward trend and fluctuates instead, further indicating a lack of convergence.

These results suggest that the model likely predicts the majority class (`neutral`) across most examples, without learning to distinguish between minority classes.

Given the **limited size and imbalance** of the dataset, such behavior is expected. To improve class sensitivity and encourage learning from underrepresented categories (`pro`, `contra`), we implement a **class-weighted loss function** in the next step.


```python
#%pip install -U "transformers[torch]" "accelerate>=0.26.0" evaluate

import evaluate
import torch
from sklearn.utils.class_weight import compute_class_weight
from torch import nn
from transformers import (
    AutoTokenizer,
    AutoModelForSequenceClassification,
    Trainer,
    TrainingArguments
)


# 1. Load evaluation metrics
accuracy = evaluate.load("accuracy")
f1 = evaluate.load("f1")

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = np.argmax(logits, axis=-1)
    return {
        "accuracy": accuracy.compute(predictions=preds, references=labels)["accuracy"],
        "f1_macro": f1.compute(predictions=preds, references=labels, average="macro")["f1"]
    }


# 2. Compute class weights
labels = tokenized_datasets["train"]["labels"]
class_weights = compute_class_weight("balanced", classes=np.unique(labels), y=labels)
class_weights = torch.tensor(class_weights, dtype=torch.float)
print("Class weights:", class_weights)

# 3. Define custom weighted model
class WeightedBERT(nn.Module):
    def __init__(self, base_model_name, num_labels, class_weights):
        super().__init__()
        self.model = AutoModelForSequenceClassification.from_pretrained(base_model_name, num_labels=num_labels)
        self.loss_fn = nn.CrossEntropyLoss(weight=class_weights)

    def forward(self, input_ids=None, attention_mask=None, token_type_ids=None, labels=None):
        outputs = self.model(
            input_ids=input_ids,
            attention_mask=attention_mask,
            token_type_ids=token_type_ids,
            labels=None
        )
        logits = outputs.logits
        loss = self.loss_fn(logits, labels) if labels is not None else None
        return {"loss": loss, "logits": logits}

# Instantiate model
base_model_name = "bert-base-german-cased"
model = WeightedBERT(base_model_name, num_labels=3, class_weights=class_weights)


# 4. Training arguments
training_args = TrainingArguments(
    output_dir="../Data/stance_model",
    eval_strategy="epoch",
    logging_dir="../Data/logs",
    num_train_epochs=5,
    learning_rate=2e-5,
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    weight_decay=0.01
)


# 5. Launch training
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets["train"],
    eval_dataset=tokenized_datasets["test"],
    tokenizer=tokenizer,
    compute_metrics=compute_metrics
)

trainer.train()
```

**Interpretation**

The BERT model was fine-tuned using a class-weighted loss function to compensate for the class imbalance in the stance-labeled dataset. However, the results indicate that this adjustment alone did not lead to substantial improvements:

- **Accuracy** remains mostly at 0.50, with a temporary drop to 0.25 in epoch 2.
- **F1 Macro** continues to fluctuate between 0.13 and 0.22, well below the random baseline of ~0.33 expected for a balanced 3-class setting.
- **Validation loss** shows minimal variation across epochs, indicating no meaningful convergence or learning dynamics.

The corresponding **confusion matrix** further confirms this trend: the model predicts the class `neutral` for **every** instance in the test set, regardless of the true label. No predictions are made for `pro` or `contra`. This behavior results in a deceptively stable accuracy, but it reflects a **majority class bias** rather than true learning.

**Conclusion**

These findings suggest that the model fails to capture stance-specific patterns in the input data. Two plausible explanations emerge:

1. **Class imbalance and limited data**: The small size of the dataset, combined with highly unequal class distributions, likely prevents the model from learning meaningful distinctions, especially for the underrepresented `pro` and `contra` classes.

2. **Content neutrality in news articles**: The model’s default to `neutral` might not only reflect training bias but also the underlying nature of the texts. As observed in the initial sentiment analysis, many newspaper articles follow a formal, fact-based tone, potentially lacking the explicit evaluative language necessary for stance detection.

### 3.3 Classical Baseline and Interpretable Models

In this new approach, we leverage the fully preprocessed texts (stored in the "clean_text" column) to predict the outcome of the referendum. Unlike previous models that focused on sentiment or stance detection, our goal here is to assess whether the underlying linguistic patterns in the cleaned articles can statistically signal the final vote outcome (e.g., "angenommen" vs. "abgelehnt").

**Workflow Overview:**
- **Feature Extraction:** We vectorize the "clean_text" using classical techniques such as TF-IDF, which transforms the text into numerical features while reducing vocabulary sparsity through prior steps like compound splitting and lemmatization.
- **Model Training:** With the extracted features, we train a supervised classifier (logistic regression) to distinguish between successful and unsuccessful initiatives.
- **Validation:** The performance of the predictive model is evaluated using appropriate metrics (accuracy, F1 score) to determine if the media narrative captured in the cleaned text aligns with the actual referendum outcomes.

This approach not only complements our descriptive and topic modelling analyses but also provides a quantifiable link between textual characteristics and public vote behavior.

#### 3.3.1 Baseline and Logisitic Regression Classifier



To establish a predictive benchmark, we compare a simple baseline model with a logistic regression classifier. As a baseline, we use a `DummyClassifier` that always predicts the majority class. For the actual classification, we train a logistic regression model on TF-IDF features derived from the cleaned article texts.

The performance of both models is evaluated using 5-fold stratified cross-validation. This setup allows us to assess whether the cleaned textual data contains enough signal to predict whether an initiative was accepted or rejected.

```python
from sklearn.dummy import DummyClassifier
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

# prepare X and y
df = pd.read_excel("../Data/Data_ML_cleaned.xlsx")
df = df[df["clean_text"].notna() & df["result"].notna()].copy()
df["label"] = df["result"].map({"Abgelehnt": 0, "Angenommen": 1})
X = df["clean_text"]
y = df["label"]

# vectorize
vectorizer = TfidfVectorizer(max_features=5000, stop_words="english")
X_tfidf = vectorizer.fit_transform(X)

# 5-fold splitter
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# baseline: always predict the majority class
dummy = DummyClassifier(strategy="most_frequent")
baseline_scores = cross_val_score(dummy, X_tfidf, y, cv=cv, scoring="accuracy")

# logistic regression
logreg = LogisticRegression(max_iter=1000)
clf_scores = cross_val_score(logreg, X_tfidf, y, cv=cv, scoring="accuracy")

print(f"Baseline accuracy (most frequent): {baseline_scores.mean():.3f} ± {baseline_scores.std():.3f}")
print(f"Logistic CV accuracy:              {clf_scores.mean():.3f} ± {clf_scores.std():.3f}")
```

**Interpretation**

At the article level, our TF-IDF + logistic regression model outperforms a naïve “always predict rejection” strategy by roughly 1.8 percentage points (a ~7 % relative error reduction). This tells us that, within our corpus, text features do capture some signal correlated with “accepted” versus “rejected” outcomes.

*Key limitation*
Because we’re randomly splitting articles and not initiatives (due to sample size) into folds, the model can simply “memorize” keywords or phrases that are specific to a given referendum. For example, if “ahv” appears in four out of five articles about the same initiative, the model will see that term in training and then easily pick it up in the held-out article. In other words, this evaluation measures how well the model recognizes already-seen campaigns, not how it would generalize to an entirely new initiative, therby limiting the interpretability.

#### 3.3.2 Cross-Validation Performance and Learning Curve

To gain a more detailed understanding of our logistic regression model’s performance, we extend the basic accuracy check in several ways. First, we use cross_validate to compute not only accuracy but also precision, recall, and F1‐score under our 5-fold split. We then summarize these results in a small table (mean ± standard deviation) and visualize them with error bars to see which metrics are most stable.

Next, we plot a learning curve (training vs. validation accuracy as we increase the sample size) to diagnose under- or over-fitting. A narrow gap between the two curves suggests good generalization, while a wide gap indicates potential over- or under-fitting.

Finally, we aggregate the 5 folds to produce an overall ROC curve—showing true positive rate vs. false positive rate across thresholds—and a confusion matrix at the 0.5 cutoff. These visualizations help us assess model discrimination and the trade-off between sensitivity and specificity, as well as the raw counts of correct vs. incorrect predictions for each class.

```python

from sklearn.model_selection import cross_validate, cross_val_predict, learning_curve
from sklearn.metrics import precision_score, recall_score, f1_score, classification_report
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, RocCurveDisplay

# 1) Cross‐validate with multiple metrics
scoring = ['accuracy', 'precision', 'recall', 'f1']
cv_results = cross_validate(logreg, X_tfidf, y, cv=cv, scoring=scoring, return_train_score=False)

# 2) Summarize mean ± std
metrics = {m: (cv_results[f'test_{m}'].mean(), cv_results[f'test_{m}'].std()) for m in scoring}
metrics_df = pd.DataFrame(metrics, index=['mean','std']).T
print(metrics_df)

# 3) Plot bar chart of mean metrics with error bars
fig, ax = plt.subplots(figsize=(8,4))
metrics_df['mean'].plot.bar(yerr=metrics_df['std'], capsize=4, ax=ax, color='skyblue')
ax.set_ylim(0,1)
ax.set_ylabel('Score')
ax.set_title('5-Fold CV Performance Metrics')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# 4) Learning Curve
train_sizes, train_scores, test_scores = learning_curve(
    logreg, X_tfidf, y, cv=cv,
    train_sizes=np.linspace(0.1,1.0,5),
    scoring='accuracy', n_jobs=-1
)
train_mean = train_scores.mean(axis=1)
test_mean  = test_scores.mean(axis=1)

fig, ax = plt.subplots(figsize=(8,4))
ax.plot(train_sizes, train_mean, 'o-', label='Train')
ax.plot(train_sizes, test_mean, 'o-', label='Validation')
ax.set_title('Learning Curve (Accuracy)')
ax.set_xlabel('Training Set Size')
ax.set_ylabel('Accuracy')
ax.legend()
ax.grid(True)
plt.tight_layout()
plt.show()

# 5) ROC Curve (aggregated folds)
y_prob = cross_val_predict(logreg, X_tfidf, y, cv=cv, method='predict_proba')[:,1]
RocCurveDisplay.from_predictions(y, y_prob)
plt.title('ROC Curve (Aggregated)')
plt.tight_layout()
plt.show()

# 6) Confusion Matrix (aggregated)
y_pred = (y_prob >= 0.5).astype(int)
cm = confusion_matrix(y, y_pred)
disp = ConfusionMatrixDisplay(cm, display_labels=['Abgelehnt','Angenommen'])
fig, ax = plt.subplots(figsize=(5,5))
disp.plot(cmap='Blues', ax=ax)
ax.set_title('Confusion Matrix (Aggregated)')
plt.tight_layout()
plt.show()
```

**Interpretation**

*Figure 1*: Cross‐Validated Performance Metrics
The bar chart shows that our logistic‐regression model achieves a overall accuracy of about 0.75 (±0.02) under 5-fold CV. Precision (≈0.40) and F1-score (≈0.11) both exhibit huge variability, and recall is near zero (≈0.06). In other words, the model is almost never correctly identifying the “accepted” (“Angenommen”) initiatives, even though it does very reliably predict the majority “rejected” (“Abgelehnt”) class.

*Figure 2*: Learning Curve (Accuracy)
As we increase the training set from 10% to 100% of the data, training accuracy falls from ~0.97 to ~0.81, while validation accuracy rises from ~0.56 to a plateau around ~0.75. The narrowing gap between the two curves suggests that adding more data improves generalization and that the model is not severely overfitting—but even with all available examples, validation accuracy levels off, indicating diminishing returns from simply feeding in more of the same data.

*Figure 3*: Aggregated ROC Curve
The aggregated ROC curve delivers an AUC of 0.98, indicating that the model’s scoring function is highly effective at ordering “accepted” vs. “rejected” articles. Yet despite this excellent ranking ability, we observe a near-zero recall in our operational evaluation—showing that, in practice, the classifier rarely labels positives correctly at its current operating point. 

*Figure 4*: Aggregated Confusion Matrix
At the 0.5 cutoff, the model correctly labels all 80 “rejected” cases and zero “rejected” false alarms, but it only catches 2 of 29 “accepted” initiatives (with 27 misses). This extreme class skew in the predictions underlines why precision, recall, and F1 suffer so badly for the positive class—even though overall accuracy and AUC look strong, the model is functionally useless for spotting the minority “accepted” outcomes.

*Conclusion:*

The model currently over-prioritizes the majority class.

Despite a high AUC, recall for “Angenommen” is effectively zero at the standard threshold.

#### 3.3.3 Feature Inspection via Logistic Regression Coefficients

To better understand which textual patterns drive the model's decisions, we now inspect the learned coefficients of the logistic regression classifier. Since TF-IDF features are interpretable by design, the sign and magnitude of each coefficient reveal how strongly a term contributes to predicting either `accepted` or `rejected` outcomes.

By fitting the model on the full dataset and extracting the top 20 positive and negative weights, we can identify the most influential words for each class.

```python
# 1) Fit on full data to extract coefficients
logreg.fit(X_tfidf, y)
feat_names = vectorizer.get_feature_names_out()
coefs      = logreg.coef_[0]

# 2) Build DataFrame of features & coefficients
feat_df = pd.DataFrame({
    'feature': feat_names,
    'coef':    coefs
})
top_pos = feat_df.nlargest(20, 'coef')
top_neg = feat_df.nsmallest(20, 'coef')

# 3) Display top drivers
#print("Top predictors for 'Angenommen':")
#display(top_pos)
#print("Top predictors for 'Abgelehnt':")
#display(top_neg)

# 4) Plot bar charts
fig, axes = plt.subplots(1,2, figsize=(12,6))
sns.barplot(data=top_pos, y='feature', x='coef', color='green', ax=axes[0])
axes[0].set_title("Drives → Angenommen")
sns.barplot(data=top_neg, y='feature', x='coef', color='red', ax=axes[1])
axes[1].set_title("Drives → Abgelehnt")
plt.tight_layout()
plt.show()

```

#### 3.3.4 Ensemble Models and Permutation-Based Feature Importance

In this section we leverage more advanced ensemble methods for outcome prediction by applying two classifiers: HistGradientBoostingClassifier and RandomForestClassifier. First, we convert the sparse TF-IDF features into a dense array (necessary for some models) and then evaluate each classifier using cross-validation. The code prints out the mean and standard deviation of accuracy scores for both models. Next, the gradient boosting model is fitted on the entire dataset and permutation importance is computed to identify which features contribute most to prediction accuracy. Finally, a bar plot is generated to visualize the top 20 feature importances. This workflow provides insights into the key drivers behind the model's decisions based on the text data.

```python

from sklearn.ensemble import HistGradientBoostingClassifier, RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.inspection import permutation_importance
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# ── Convert sparse TF-IDF to dense ──
X_dense = X_tfidf.toarray()

# 1) More advanced classifiers
gb = HistGradientBoostingClassifier(
    loss='log_loss',
    learning_rate=0.05,
    max_iter=200,
    random_state=42
)
rf = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    n_jobs=-1
)

# 2) CV accuracy
gb_scores = cross_val_score(gb, X_dense, y, cv=cv, scoring='accuracy')
rf_scores = cross_val_score(rf, X_tfidf, y, cv=cv, scoring='accuracy')
print(f"HistGB accuracy:      {gb_scores.mean():.3f} ± {gb_scores.std():.3f}")
print(f"RandomForest accuracy:{rf_scores.mean():.3f} ± {rf_scores.std():.3f}")

# 3) Fit GB on full data
gb.fit(X_dense, y)

# 4) Permutation importances (works even if feature_importances_ is missing)
perm = permutation_importance(
    gb, X_dense, y,
    n_repeats=10,
    random_state=42,
    n_jobs=-1
)
feat_names = vectorizer.get_feature_names_out()
importances = perm.importances_mean
feat_imp = pd.Series(importances, index=feat_names).nlargest(20)

# 5) Plot top 20 importances
plt.figure(figsize=(8,6))
sns.barplot(x=feat_imp.values, y=feat_imp.index, color='purple')
plt.title("Top 20 Feature Importances – HistGradientBoosting (Perm. Imp.)")
plt.xlabel("Mean decrease in accuracy")
plt.ylabel("Feature")
plt.tight_layout()
plt.show()
```

**Interpretation**

- **Accuracy**: Both HistGradientBoosting and RandomForest reach 0.744±0.06/0.744±0.03 in article-level 5-fold CV, i.e. no real gain over logistic regression.  
- **Top features**: “arbeits” (–0.054), “steuer” (–0.020), “stark” (–0.008), “leben” (–0.007) drive the model.  
This means that the majority of the other features exhibit near-zero permutation importance. They are considered noise by the model.


*Conclusions*

Our article-level benchmark shows a modest lift—from ~0.73 (dummy) to ~0.75 accuracy—when moving from a majority‐class baseline to logistic regression or tree ensembles. However, because we randomly split **articles** rather than **initiatives**, the model can simply memorize campaign-specific keywords, inflating both accuracy and feature‐importance signals.
In the broader workflow—alongside our descriptive statistics and topic‐modeling analyses—this exercise demonstrates that **some** linguistic signal exists, but true out-of-sample generalization to **new** referenda requires more data.

### Final Interpretation & Conclusion

Taken together, the results of this exploratory analysis suggest that media coverage of Swiss referenda tends to be neutral in tone, factual in language, and institutionally consistent across initiatives. Neither framing theory nor mirror theory is strongly supported by the findings.

- **Topic and word frequency analysis**: revealed some thematic patterns: socially oriented terms (e.g., *pflegen*, *AHV*, *frau*) were more common in accepted initiatives, whereas rejected ones often included more institutional, economic, or regulatory vocabulary (e.g., *steuer*, *unternehmen*, *wirtschaft*). However, these findings are **descriptive** and cannot determine causal framing effects.

- **Sentiment and stance detection**: Sentiment analysis classifies nearly all articles as neutral, and stance detection fails to extract clear positions (even after applying a class-weighted loss function), indicating that articles rarely adopt overtly evaluative language. However, result may also stem form unbalanced and small dataset as it was based on 40 pre-classified articles.

- **Outcome prediction models**: based on TF-IDF and logistic regression marginally outperformed a naïve baseline. Tree-based models (Random Forest, HistGradientBoosting) showed no substantial improvement and confirmed that only a handful of features held predictive value. Crucially, many of the most informative terms appear to be **highly initiative-specific**, raising the concern that the models may be learning to recognize known campaigns rather than generalizable patterns.

- Most models were evaluated on **article-level random splits**, not **initiative-level splits**, which further limits interpretability. This setup risks data leakage by allowing campaign-specific language to appear in both train and test sets, artificially inflating performance metrics and feature importances.

Taken together, these findings point to a **general tendency toward neutral and fact-oriented reporting** in Swiss political journalism. This overall neutrality makes it difficult to identify consistent or generalizable links between media language and vote outcomes.  
**However, this observed neutrality may also be shaped by the small and imbalanced dataset, which could obscure more nuanced forms of framing or variation between media outlets.**

Rather than strongly reflecting or shaping public preferences, media coverage in this context appears to follow a more institutional, impartial reporting style. While some textual patterns may align with eventual outcomes, they are not robust enough to support systematic prediction or theoretical generalization.

To robustly investigate framing effects and media representativeness in democratic contexts, future work should prioritize:
- Larger and more balanced datasets,
- Initiative-level or time-based cross-validation
